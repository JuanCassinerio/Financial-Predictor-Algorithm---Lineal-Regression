```python
#Analisis de variables

plt.figure(figsize=(9,3))
plt.title("stock price growth vs gdp growth")
plt.xlabel('gdp growth')
plt.ylabel('stock price growth')
sns.scatterplot(data=df,x='gdpc',y='sc')
plt.xlim(0,0.05)
plt.show()


plt.figure(figsize=(9,3))
plt.title("stock price growth vs gdp growth")
plt.xlabel('Ocupacion')
plt.ylabel('Subscripcion a Deposito')
sns.scatterplot(data=df,x='i',y='sc')
plt.show()
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
x=df[['ec','i','gdpc','ffr']]
y=df[['sc']]
linear=LinearRegression()
model=linear.fit(x,y)

r_sq = model.score(x, y)
print(f"accuracy: {r_sq}")
```

```python
x_pred=[[-0.03442,0.047,0.0567,0.0008]]
y_pred=model.predict(x_pred)
y_pred
```
